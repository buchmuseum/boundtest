```python
html_content = """<!DOCTYPE html>
<html lang="de">
<head>
  <meta charset="UTF-8">

  <!-- 🔥 WICHTIG für Smartphone -->
  <meta name="viewport" content="width=device-width, initial-scale=1.0">

  <title>Leipzig Bound</title>

  <style>
    body {
      font-family: sans-serif;
      text-align: center;
      padding: 10px;
      margin: 0;
      background: #f5f5f5;
    }

    #app {
      max-width: 600px;
      margin: auto;
      background: white;
      padding: 70px 15px 15px 15px; /* nur einmal sauber */
      border-radius: 10px;
    }

    #hintBtn {
      position: fixed;
      top: 10px;
      right: 10px;
      z-index: 1000;
      padding: 10px 14px;
      font-size: 16px;
      background: white;
      border: 1px solid #ccc;
      border-radius: 8px;
      box-shadow: 0 2px 6px rgba(0,0,0,0.1);
    }

    button {
      display: block;
      margin: 10px auto;
      padding: 14px;
      width: 100%;
      font-size: 18px;
      border-radius: 8px;
    }

    input[type="text"] {
      font-size: 18px;
      padding: 10px;
    }

    button.correct-selected {
      background: #cfffe6 !important;
    }

    button.wrong-selected {
      background: #ffb3b3 !important;
    }

    .check-btn {
      background: #2d6cdf;
      color: white;
      font-weight: bold;
      border: none;
    }

    .check-btn:hover {
    background: #1f4fb3;
    }

    img {
      max-width: 100%;
      height: auto;
      border-radius: 8px;
    }
  </style>
</head>

<body>

<button id="hintBtn" style="position:fixed; top:10px; right:10px; padding:10px 14px; font-size:16px;">
💡 Möchtest du einen Tipp?
</button>

<div id="hintBox" style="display:none; position:fixed; bottom:10px; left:10px; right:10px; background:#fff3cd; border:1px solid #ffeeba; padding:12px; border-radius:8px; font-size:16px;">
</div>

<div id="app"></div>




<script>
let current = 0;
let score = 0;

let selected = [];
let attempts = 0;
let locked = false;
let currentHint = null;


const hintBtn = document.getElementById("hintBtn");
const hintBox = document.getElementById("hintBox");


if (currentHint) {
  hintBtn.style.display = "block";
} else {
  hintBtn.style.display = "none";
  hintBox.style.display = "none";
}

hintBtn.onclick = () => {
  if (!currentHint) return;

  hintBox.style.display = "block";
  hintBox.innerHTML = "💡 Tipp: " + currentHint;
};

const data = [
  {
    type: "info",
    text: `<h2>Schön, dass du da bist!</h2> 
    
In diesem spannenden Actionbound-Quiz gehst du auf Entdeckungsreise durch unsere Dauerausstellung. 
🕵️‍♀️ Du wirst faszinierende Dinge über Bücher, das Schreiben und alte Schriften entdecken – von Federkiel bis Digitalzeit. 
✍️📱 Wusstest du, dass das Smartphone in deiner Hand eine ganz moderne Fortsetzung jahrtausendealter Schreibtraditionen ist? 
⏱️ Keine Sorge: Deine Reise dauert nur etwa 30 Minuten. Los geht’s – viel Spaß!`
  },

  {
    type: "quiz",
    text: `📜 Wie alles begann – ganz ohne Schrift!`,
    image: "2.jpg",
    question: "Wofür wurden die Ketten benutzt?",
    answers: [
      "Schmuck",
      "Legitimation von Botschaften",
      "Erinnerungsstütze",
      "Währung",
      "Rituale"
    ],
    correct: [1, 2],
    hint: "2 Antworten sind korrekt"
  },

  {
    type: "info",
    text: `🗿 Geschichten aus Stein? 
    <br>
    <br>
    Ja, wirklich! Bevor es Bücher gab, haben Menschen ihre Geschichten ganz anders erzählt. <br>
    Im Eingangsbereich des Museums seht ihr einen assyrischen Obelisken! <br>
    In diesen Stein wurden Bilder und Zeichen eingeritzt – fast wie ein Comic aus der Steinzeit!`
  },

  {
    type: "quiz",
    text: `Woher kommt eigentlich unser Alphabet?

   Das Alphabet, das wir heute benutzen, ist schon richtig alt – und es hat eine spannende Geschichte! 
  📚 Wir alle lernen die Buchstaben in der Schule und benutzen sie jeden Tag – beim Lesen, Schreiben oder Chatten.

  Aber wusstest du, dass es viele, viele Jahre gebraucht hat, bis unser Alphabet so aussah wie heute? 
  🤔 Hast du eine Idee, wo es ursprünglich herkommt?`,
    question: "Woher kommt unser Alphabet ursprünglich?",
    answers: [
      "Von den Griechen",
      "Von den Ägyptern",
      "Von den Phöniziern",
      "Von den Maya"
    ],
    correct: [2]
  },

  {
    type: "ltext",
    text: `🌟 Die erste „Himmelskarte“!

Bevor es Karten und Kalender gab, mussten die Menschen sich anders behelfen. Eines dieser alten „Zeitdokumente“ ist die Himmelsscheibe von Nebra, die ihr in der Vitrine vor euch seht. 🪐

Sie ist eine der ersten Darstellungen des Sternenhimmels und wurde auch als Kalender genutzt! 
📅 Auf der Scheibe kann man die Sonne, den Mond und viele Sterne sehen.`,
    question: "Kannst du zählen, wie viele Sterne insgesamt abgebildet sind?",
    correctText: "30"
  },

  {
  type: "cloze_letters",
  text: `Wo werden diese Schätze in Leipzig gesammelt? : Deutsches ____ - und ____museum`,
  answers: ["Buch", "Schrift"]
  },

  {
    type: "timeline",
    text: "Wann wurde der Buchdruck erfunden?",
    question: "Schätze das Jahr:",
    min: 0,
    max: 2000,
    correct: 1440,
    tolerance: 100
  },

  
  {
    type: "sort",
    question: "Sortiere die Entwicklung der Schrift:",
    items: [
      "Höhlenmalerei",
      "Hieroglyphen",
      "Alphabet",
      "Buchdruck"
    ],
    correct: [0, 1, 2, 3]
  },
  {
  type: "match",
  question: "Ordne die Kulturen ihren Schriftsystemen zu:",
  pairs: [
    ["Ägypten", "Hieroglyphen"],
    ["Griechen", "Alphabet"],
    ["China", "Schriftzeichen"]
  ]
  },
  {
  type: "multi_select",
  question: "Welche dieser Dinge gehören zur Geschichte der Schrift?",
  answers: [
    "Papyrus",
    "Pergament",
    "Steinritzungen",
    "Druck",
    "Plastik",
    "Smartphone",
    "Hieroglyphen",
    "Alphabet",
    "Emoji"
  ],
  correct: [0, 1, 2, 3]
}

];

function resetState() {
  selected = [];
  attempts = 0;
  locked = false;
}

function evaluate(item) {
  if (item.type !== "quiz") return "wrong";

  const correct = item.correct;

  const hasWrong = selected.some(i => !correct.includes(i));
  const missingCorrect = correct.some(i => !selected.includes(i));

  if (missingCorrect) return "wrong";
  if (hasWrong) return "partial";
  return "perfect";
}

function getMaxAttempts(item) {
  if (item.type === "quiz") return item.correct.length + 1;
  if (item.type === "ltext") return 2;
  if (item.type === "timeline") return 2;
  return 1;
}

function next() {
  current++;
  resetState();
  render();
}

function showEnd() {
  app.innerHTML = `
    <h2>🎉 Fertig!</h2>
    <p>Punkte: ${score}</p>
  `;
}

function showFeedback(result, points) {
  let msg = "";

  if (result === "perfect") msg = "🟢 Perfekt!";
  if (result === "partial") msg = "🟡 Ein kleiner Fehler war dabei, daher halbe Punktzahl";

  app.innerHTML = `
    <h2>${msg}</h2>
    <p>Punkte: ${points}</p>
    <button onclick="next()">Weiter</button>
  `;
}

function showSolution(item) {
  let content = "<h2>❌ Lösung</h2>";

  if (item.type === "quiz") {
    content += "<p>Richtig waren:</p><ul>";
    content += item.answers.map((a, i) =>
      item.correct.includes(i) ? `<li>✅ ${a}</li>` : ""
    ).join("");
    content += "</ul>";
  }

  else if (item.type === "ltext") {
    content += `<p>Richtige Antwort: <b>${item.correctText}</b></p>`;
  }

  else if (item.type === "timeline") {
    content += `<p>Richtiges Jahr: <b>${item.correct}</b></p>`;
  }
  else if (item.type === "sort") {
  content += "<p>Richtige Reihenfolge:</p><ol>";

  const correctOrder = item.correct.map(i => item.items[i]);

  content += correctOrder.map(item => `<li>✅ ${item}</li>`).join("");

  content += "</ol>";
}

  content += `<button onclick="next()">Weiter</button>`;
  app.innerHTML = content;
}

function shuffle(array) {
  let arr = [...array];
  for (let i = arr.length - 1; i > 0; i--) {
    let j = Math.floor(Math.random() * (i + 1));
    [arr[i], arr[j]] = [arr[j], arr[i]];
  }
  return arr;
}

function render() {
  if (!data[current]) {
    showEnd();
    return;
  }

  app.innerHTML = "";

  const item = data[current];

  currentHint = item.hint || null;

  const hintBtn = document.getElementById("hintBtn");
  const hintBox = document.getElementById("hintBox");

  if (currentHint) {
    hintBtn.style.display = "block";
  } else {
    hintBtn.style.display = "none";
    hintBox.style.display = "none";
  }

  const container = document.createElement("div");

  if (item.text && item.type !== "cloze_letters") {
    const t = document.createElement("p");
    t.innerHTML = item.text;
    container.appendChild(t);
  }

  if (item.image) {
    const img = document.createElement("img");
    img.src = item.image;
    img.style.maxWidth = "100%";
    container.appendChild(img);
  }

  if (item.type === "info") {
    const btn = document.createElement("button");
    btn.textContent = "Weiter";
    btn.onclick = next;
    container.appendChild(btn);
  }

  
  // MultSelect
  if (item.type === "multi_select") {
  const q = document.createElement("h2");
  q.textContent = item.question;
  container.appendChild(q);

  let selected = [];
  let attempts = 0;
  const maxAttempts = 2;

  item.answers.forEach((text, index) => {
    const btn = document.createElement("button");
    btn.textContent = text;

    btn.onclick = () => {
      if (locked) return;

      // toggle selection
      if (selected.includes(index)) {
        selected = selected.filter(i => i !== index);
        btn.style.background = "";
      } else {
        selected.push(index);
        btn.style.background = "#cfe8ff";
      }
    };

    container.appendChild(btn);
  });

  const check = document.createElement("button");
  check.textContent = "Prüfen";

  check.onclick = () => {
    if (locked) return;

    attempts++;

    const correctSet = item.correct;

    const isPerfect =
      selected.length === correctSet.length &&
      selected.every(i => correctSet.includes(i));

    const hasSomeCorrect = selected.some(i => correctSet.includes(i));

    // 🟢 perfekt
    if (isPerfect) {
      locked = true;

      if (attempts === 1) {
        score += 1;
        showFeedback("perfect", 1);
      } else {
        score += 0.5;
        showFeedback("partial", 0.5);
      }

      return;
    }

    // 🟡 teilweise richtig
    if (hasSomeCorrect && attempts < maxAttempts) {
      alert("🟡 Deine Lösungen stimmten leider nicht alle – versuch es nochmal!");
      return;
    }

    // ❌ falsch oder letzter Versuch
    if (attempts < maxAttempts) {
      alert("❌ Das ist leider Falsch – letzter Versuch!");
      return;
    }

    locked = true;
    showSolution(item);
  };

  container.appendChild(check);
}

  // CLOZE LETTERS (Buchstaben einzeln)
if (item.type === "cloze_letters") {

  let attempts = 0;
  const maxAttempts = 2;

  const parts = item.text.split("____");
  const wrapper = document.createElement("p");

  const inputs = [];
  const inputList = [];

  // wichtig: verhindert alte Inputs aus vorherigen Fragen
  inputList.length = 0;

  parts.forEach((part, i) => {
    wrapper.appendChild(document.createTextNode(part));

    if (i < item.answers.length) {
      const word = item.answers[i];
      const wordContainer = document.createElement("span");

      for (let j = 0; j < word.length; j++) {
        const input = document.createElement("input");
        input.type = "text";
        input.maxLength = 1;

        input.style.width = "25px";
        input.style.margin = "2px";
        input.style.textAlign = "center";

        // speichern für Prüfung + Navigation
        inputs.push({ input, correct: word[j].toLowerCase() });
        inputList.push(input);

        // 👉 AUTO NEXT
        input.addEventListener("input", () => {
          input.value = input.value.slice(0, 1);

          const index = inputList.indexOf(input);

          if (input.value && inputList[index + 1]) {
            inputList[index + 1].focus();
          }
        });

        // 👉 BACKSPACE ZURÜCK
        input.addEventListener("keydown", (e) => {
          const index = inputList.indexOf(input);

          if (e.key === "Backspace" && !input.value && inputList[index - 1]) {
            inputList[index - 1].focus();
          }
        });

        wordContainer.appendChild(input);
      }

      wrapper.appendChild(wordContainer);
    }
  });

  container.appendChild(wrapper);

  // 👉 CHECK BUTTON
  const btn = document.createElement("button");
  btn.textContent = "Prüfen";
  btn.className = "check-btn";

  btn.onclick = () => {
    if (locked) return;

    attempts++;

    let correctCount = 0;

    inputs.forEach(({ input, correct }) => {
      const val = input.value.toLowerCase();

      if (val === correct) {
        input.style.background = "#cfe8ff";
        correctCount++;
      } else {
        input.style.background = "#ffb3b3";
      }
    });

    if (correctCount === inputs.length) {
      locked = true;

      if (attempts === 1) {
        score += 1;
        showFeedback("perfect", 1);
      } else {
        score += 0.5;
        showFeedback("partial", 0.5);
      }

      return;
    }

    if (attempts < maxAttempts) {
      alert("❌ Deine Lösung ist leider nicht richtig – versuch es nochmal!");
      return;
    }

    locked = true;

    let solution = "<p>Richtige Lösung:</p><ul>";
    item.answers.forEach(a => {
      solution += `<li>${a}</li>`;
    });
    solution += "</ul>";

    app.innerHTML = `
      <h2>❌ Leider falsch, die Lösung ist:</h2>
      ${solution}
      <button onclick="next()">Weiter</button>
    `;
  };

  container.appendChild(btn);
}

  // LTEXT
  if (item.type === "ltext") {
    const q = document.createElement("h2");
    q.textContent = item.question;
    container.appendChild(q);

    const input = document.createElement("input");
    input.type = "text";
    input.style.width = "80%";
    container.appendChild(input);

    let ltextAttempts = 0;

    const btn = document.createElement("button");
    btn.textContent = "Prüfen";
    btn.className = "check-btn";

    btn.onclick = () => {
      if (locked) return;

      ltextAttempts++;

      if (input.value.trim().toLowerCase() === item.correctText.toLowerCase()) {
        locked = true;
        score += 1;
        showFeedback("perfect", 1);
        return;
      }

      if (ltextAttempts >= getMaxAttempts(item)) {
        locked = true;
        showSolution(item);
      } else {
        alert("❌ Deine Lösung ist leider nicht richtig – noch ein Versuch!");
      }
    };

    container.appendChild(btn);
  }

  // TIMELINE
  if (item.type === "timeline") {
    const q = document.createElement("h2");
    q.textContent = item.question;
    container.appendChild(q);

    const input = document.createElement("input");
    input.type = "range";
    input.min = item.min;
    input.max = item.max;
    input.value = item.min;

    const display = document.createElement("p");
    display.textContent = input.value;

    input.oninput = () => display.textContent = input.value;

    const btn = document.createElement("button");
    btn.textContent = "Prüfen";
    btn.className = "check-btn";

    btn.onclick = () => {
      if (locked) return;

      const diff = Math.abs(parseInt(input.value) - item.correct);
      locked = true;

      if (diff === 0) {
        score += 1;
        showFeedback("perfect", 1);
      } else if (diff <= item.tolerance) {
        score += 0.5;
        showFeedback("partial", 0.5);
      } else {
        showSolution(item);
      }
    };

    container.appendChild(q);
    container.appendChild(display);
    container.appendChild(input);
    container.appendChild(btn);
  }

  // MATCH (DRAG & DROP VERSION)
if (item.type === "match") {

  const q = document.createElement("h2");
  q.textContent = item.question;
  container.appendChild(q);

  const leftCol = document.createElement("div");
  const rightCol = document.createElement("div");

  leftCol.style.display = "inline-block";
  rightCol.style.display = "inline-block";
  rightCol.style.marginLeft = "20px";

  container.appendChild(leftCol);
  container.appendChild(rightCol);

  let selectedLeft = null;
  let matches = {};
  let attempts = 0;
  const maxAttempts = 2;

  const colors = ["#cfe8ff", "#e6ccff", "#d6f5d6"];
  let colorIndex = 0;

  const leftButtons = [];
  const rightButtons = [];

  // LINKS
  item.pairs.forEach(([left, right]) => {

    const btn = document.createElement("button");
    btn.textContent = left;
    btn.className = "match-left";

    btn.onclick = () => {
      if (locked) return;

      selectedLeft = left;

      leftButtons.forEach(b => b.style.background = "");

      btn.style.background = "#cfe8ff";
    };

    leftCol.appendChild(btn);
    leftButtons.push(btn);
  });

  // RECHTS
  const shuffled = [...item.pairs.map(p => p[1])]
    .sort(() => Math.random() - 0.5);

  shuffled.forEach(right => {

    const btn = document.createElement("button");
    btn.textContent = right;
    btn.className = "match-right";

    btn.onclick = () => {
      if (locked) return;
      if (!selectedLeft) {
        alert("Bitte zuerst links auswählen");
        return;
      }

      const color = colors[colorIndex++ % colors.length];

      matches[selectedLeft] = right;

      // links einfärben
      leftButtons.forEach(b => {
        if (b.textContent === selectedLeft) {
          b.style.background = color;
        }
      });

      // rechts einfärben
      btn.style.background = color;

      selectedLeft = null;
    };

    rightCol.appendChild(btn);
    rightButtons.push(btn);
  });

  // CHECK
  const check = document.createElement("button");
  check.textContent = "Prüfen";

  check.onclick = () => {
    if (locked) return;

    attempts++;

    let correct = 0;

    item.pairs.forEach(([l, r]) => {
      if (matches[l] === r) correct++;
    });

    if (correct === item.pairs.length) {
      locked = true;
      score += attempts === 1 ? 1 : 0.5;
      showFeedback(attempts === 1 ? "perfect" : "partial", 1);
      return;
    }

    if (attempts < maxAttempts) {
      alert("❌ Deine Lösung ist leider nicht richtig - versuch es nochmal!");
      return;
    }

    locked = true;
    showSolution(item);
  };

  container.appendChild(check);
}


   // SORT (DRAG & DROP)
if (item.type === "sort") {

  const q = document.createElement("h2");
  q.textContent = item.question;
  container.appendChild(q);

  let order = shuffle([...item.items]);
  let dragged = null;
  let attempts = 0;
  const maxAttempts = 2;

  const list = document.createElement("div");

  function renderList() {
    list.innerHTML = "";

    order.forEach((text, index) => {
      const row = document.createElement("div");

      row.textContent = text;
      row.style.padding = "12px";
      row.style.margin = "6px 0";
      row.style.background = "#f0f0f0";
      row.style.borderRadius = "8px";
      row.style.cursor = "grab";

      row.draggable = true;

      // 🔥 DRAG START
      row.ondragstart = () => {
        dragged = index;
      };

      // 🔥 DROP TARGET
      row.ondragover = (e) => {
        e.preventDefault();
      };

      row.ondrop = () => {
        if (dragged === null) return;

        const temp = order[dragged];
        order.splice(dragged, 1);
        order.splice(index, 0, temp);

        dragged = null;
        renderList();
      };

      list.appendChild(row);
    });
  }

  const btn = document.createElement("button");
  btn.textContent = "Prüfen";

  btn.onclick = () => {

    if (locked) return;

    attempts++;

    const correctOrder = item.correct.map(i => item.items[i]);

    const isCorrect = JSON.stringify(order) === JSON.stringify(correctOrder);

    if (isCorrect) {
      locked = true;

      if (attempts === 1) {
        score += 1;
        showFeedback("perfect", 1);
      } else {
        score += 0.5;
        showFeedback("partial", 0.5);
      }
      return;
    }

    if (attempts < maxAttempts) {
      alert("❌ Deine Lösung ist leider nicht richtig – versuch es nochmal!");
      return;
    }

    locked = true;
    showSolution(item);
  };

  container.appendChild(list);
  container.appendChild(btn);

  renderList();
}

  // QUIZ
  if (item.type === "quiz") {
    const q = document.createElement("h2");
    q.textContent = item.question;
    container.appendChild(q);

    item.answers.forEach((answer, index) => {
      const btn = document.createElement("button");
      btn.className = "quiz-btn";
      btn.textContent = answer;

      btn.onclick = () => {
      if (locked) return;

      // 👉 verhindern, dass man denselben Button nochmal klickt
      if (btn.disabled) return;

      btn.disabled = true; // 🔒 Button sperren

      selected.push(index);
      attempts++;

      const result = evaluate(item);
      const maxAttempts = getMaxAttempts(item);

      const buttons = container.querySelectorAll(".quiz-btn");

      buttons.forEach((b, i) => {
        if (selected.includes(i)) {
          if (item.correct.includes(i)) {
            b.classList.add("correct-selected");
          } else {
            b.classList.add("wrong-selected");
          }
        }
      });

      const isLastAttempt = attempts >= maxAttempts;

      if (result === "perfect") {
        locked = true;
        score += 1;
        showFeedback("perfect", 1);
        return;
      }

      if (result === "partial" && isLastAttempt) {
        locked = true;
        score += 0.5;
        showFeedback("partial", 0.5);
        return;
      }

      if (result === "wrong" && isLastAttempt) {
        locked = true;
        showSolution(item);
        return;
      }
    };

      container.appendChild(btn);
    });
  }

  app.appendChild(container);
}

render();
</script>

</body>
</html>
"""

with open("boundsmart.html", "w", encoding="utf-8") as f:
    f.write(html_content)

print("HTML gespeichert!")
```
